## Bar graph

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,4), dpi=100)
plt.bar(labels, values,color='blue')
plt.title('BAR GRAPH',fontdict={'fontname':'MS Gothic','fontsize':20})
plt.show()
```

#### Patterns in Bar graph

```python
#Method-1:

labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,4), dpi=100)

bars=plt.bar(labels, values,color='blue')

bars[0].set_hatch('/')
bars[1].set_hatch('O')
bars[2].set_hatch('*')

plt.title('BAR GRAPH',fontdict={'fontname':'MS Gothic','fontsize':20})
plt.show()
```

```python
#Method-2:

labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,4), dpi=100)

bars=plt.bar(labels, values,color='blue')

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))    #Pops the first pattern off and fits it in bars[0] and so on.

plt.savefig('barchart.png', dpi=300)

plt.title('BAR GRAPH',fontdict={'fontname':'MS Gothic','fontsize':20})
plt.show()
```

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,4), dpi=100)

bars=plt.bar(labels, values,color='blue')

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop())    #Pops the last pattern off and fits it in bars[0] and so on.

plt.savefig('barchart1.png', dpi=300)

plt.title('BAR GRAPH',fontdict={'fontname':'MS Gothic','fontsize':20})
plt.show()
```

## Real-World Examples

### Line Graph

```python
gas = pd.read_csv('gas_prices.csv.txt')

plt.figure(figsize=(8,5))

plt.title('Gas Prices over Time (in USD)', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.plot(gas.Year, gas.USA, 'b.-',label='USA')

#We can provide a label seperately to change the name of the countries if we want. In this we already have the labels of
#the countries in our gas_prices.csv file.
#plt.plot(gas.Year, gas.USA, 'b.-')

# WE HAVE TO PROVIDE LABELS IN CASE OF csv.txt FILES.

plt.plot(gas.Year, gas.Canada, 'r.-',label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-',label='South Korea')   

# We can also write the x or y-axis like gas['Year'] or gas['South Korea'].

plt.plot(gas.Year, gas.Australia, 'y.-',label='Australia')

#Here, the step size for the terms is 3 on the x-axis, you can alter it according to your needs

plt.xticks(gas.Year[::3].tolist()+[2011])  #In this if we want to add another year in the graph, then simply add '.tolist()'

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('Gas_price_figure.png', dpi=300)

plt.show()
```

```python
# Another Way to plot many values!

gas = pd.read_csv('gas_prices.csv.txt')
plt.figure(figsize=(8,5))
plt.title('Gas Prices over Time (in USD)', fontdict={'fontname':'MS Gothic','fontweight':'bold', 'fontsize': 18})


countries_to_look_at = ['Australia', 'USA', 'Canada', 'South Korea']
for country in gas:
    if country in countries_to_look_at:
         plt.plot(gas.Year, gas[country], marker='.')
            
#FOR ALL COUNTRIES:
#for country in gas:
    # if country !='Year':
        # plt.plot(gas.Year, gas[country], marker='.')
        
plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()

plt.savefig('Gas_price_figure1.png', dpi=300)

plt.show()
```

### Load Fifa Data

```python
fifa = pd.read_csv('fifa_data.csv.txt')
fifa.head(5)
```

### Histogram

```python
b = [40,50,60,70,80,90,100]   #Bins (represented by any variable, eg=b) decide the range of each value in the x-axis: 40-50,50-60 etc.

plt.figure(figsize=(8,5))

plt.hist(fifa.Overall, bins=b, color='#ad6ef0')

plt.xticks(bins)

plt.ylabel('Number of Players')
plt.xlabel('Skill Level')
plt.title('Distribution of Player Skills in FIFA 2018')

plt.savefig('histogram.png', dpi=300)

plt.show()
```

### Pie Chart

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.figure(figsize=(8,5))

label = ['Left', 'Right']
color = ['#6ef0d2', '#04bab1']

# autopct sets the percentage of left and right in the pie chart.
plt.pie([left, right], labels = label, colors=color, autopct='%.2f %%')   #%% is to write % in the pie chart

plt.title('Foot Preference of FIFA Players')

plt.show()
```

### Pie Chart 2

```python
#STYLE-1:

plt.figure(figsize=(8,5), dpi=100)

#To remove the string 'lbs' from the weigh values and convert it into integer:
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]  

# This is basically like:
# for x in fifa.Weight:
    # if type(x) == str:
        # int(x.strip('lbs'))
    #else:
        # x

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label,autopct='%.2f %%')
plt.show()
```

```python
#STYLE-2: Since the previous pie chart is not as presentable or easy to read.
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')   #You can change the color combination of the pie chart using this.

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

# 'pctdistance' is for increasing/decreasing the percentage(values) distances in the graph.
# 'explode' is for scattering different pieces of the pie chart a bit further apart for clear reading.
# Since we only need to explode 'Under 125'(.4) and 'Over 200'(.4), hence we do not explode the other parts(0)

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```

### Box and Whiskers Chart

```python
plt.style.use('default')

plt.figure(figsize=(5,8))

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

# 'patch_artist=True' is necessary for box and whiskers chart for setting the facecolor
# 'medianprops' is for changing the median linewidth

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

#Another method of changing the styles instead of directly writing it in boxplot

for box in bp['boxes']:        # ['boxes'] For indicating the changes on in the rectangular box and not the whiskers
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')
    
plt.show()
```
